```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report

import pickle
from pickle import dump

import warnings
warnings.filterwarnings('ignore')
```

```python
# Your code here
with open('/workspaces/kiliangil-Boosting-Prediciendo-diabetes/data/processed/dataset.pkl', 'rb') as f:
    X_train, X_test, y_train, y_test = pickle.load(f)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

#### Modelado - Prueba 1

```python
# modelos
ada = AdaBoostClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, random_state = 42, max_depth= 5, use_label_encoder=False, eval_metric='mlogloss')
lgb = LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state = 42)

# entrenamiento
ada.fit(X_train, y_train)
gb.fit(X_train, y_train)
xgb.fit(X_train, y_train)
lgb.fit(X_train, y_train)
```

```python
# predicción
ada_y_pred_test = ada.predict(X_test)
gb_y_pred_test = gb.predict(X_test)
xgb_y_pred_test = xgb.predict(X_test)
lgb_y_pred_test = lgb.predict(X_test)

ada_y_pred_train = ada.predict(X_train)
gb_y_pred_train = gb.predict(X_train)
xgb_y_pred_train = xgb.predict(X_train)
lgb_y_pred_train = lgb.predict(X_train)
```

```python
# metricas
ada_accuracy_test = accuracy_score(y_test, ada_y_pred_test)
ada_accuracy_train = accuracy_score(y_train, ada_y_pred_train)

gb_accuracy_test = accuracy_score(y_test, gb_y_pred_test)
gb_accuracy_train = accuracy_score(y_train, gb_y_pred_train)

xgb_accuracy_test = accuracy_score(y_test, xgb_y_pred_test)
xgb_accuracy_train = accuracy_score(y_train, xgb_y_pred_train)

lgb_accuracy_test = accuracy_score(y_test, lgb_y_pred_test)
lgb_accuracy_train = accuracy_score(y_train, lgb_y_pred_train)

print('AdaBoost')
print("Accuracy Test: ", ada_accuracy_test)
print("Accuracy Train: ", ada_accuracy_train)

print('Gradient Boosting')
print("Accuracy Test: ", gb_accuracy_test)
print("Accuracy Train: ", gb_accuracy_train)

print('XGBoost')
print("Accuracy Test: ", xgb_accuracy_test)
print("Accuracy Train: ", xgb_accuracy_train)

print('LightGBM')
print("Accuracy Test: ", lgb_accuracy_test)
print("Accuracy Train: ", lgb_accuracy_train)
```

#### Prueba 3

```python
# modelos
ada11 = AdaBoostClassifier(n_estimators=50, learning_rate=0.5, random_state=42)
gb11 = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, max_depth=2, random_state=42)
xgb11 = XGBClassifier(n_estimators=50, learning_rate=0.1, random_state = 42, max_depth= 3, colsample_bytree= 0.8, use_label_encoder=False, eval_metric='mlogloss')
lgb11 = LGBMClassifier(n_estimators=40, learning_rate=0.1, num_leaves= 20, colsample_bytree= 0.7, random_state = 42)

# entrenamiento
ada11.fit(X_train, y_train)
gb11.fit(X_train, y_train)
xgb11.fit(X_train, y_train)
lgb11.fit(X_train, y_train)
```

```python
# predicción
ada_y_pred_test11 = ada11.predict(X_test)
gb_y_pred_test11 = gb11.predict(X_test)
xgb_y_pred_test11 = xgb11.predict(X_test)
lgb_y_pred_test11 = lgb11.predict(X_test)

ada_y_pred_train11 = ada11.predict(X_train)
gb_y_pred_train11 = gb11.predict(X_train)
xgb_y_pred_train11 = xgb11.predict(X_train)
lgb_y_pred_train11 = lgb11.predict(X_train)
```

```python
# metricas
ada_accuracy_test11 = accuracy_score(y_test, ada_y_pred_test11)
ada_accuracy_train11 = accuracy_score(y_train, ada_y_pred_train11)

gb_accuracy_test11 = accuracy_score(y_test, gb_y_pred_test11)
gb_accuracy_train11 = accuracy_score(y_train, gb_y_pred_train11)

xgb_accuracy_test11 = accuracy_score(y_test, xgb_y_pred_test11)
xgb_accuracy_train11 = accuracy_score(y_train, xgb_y_pred_train11)

lgb_accuracy_test11 = accuracy_score(y_test, lgb_y_pred_test11)
lgb_accuracy_train11 = accuracy_score(y_train, lgb_y_pred_train11)

print('AdaBoost')
print("Accuracy Test: ", ada_accuracy_test11)
print("Accuracy Train: ", ada_accuracy_train11)

print('Gradient Boosting')
print("Accuracy Test: ", gb_accuracy_test11)
print("Accuracy Train: ", gb_accuracy_train11)

print('XGBoost')
print("Accuracy Test: ", xgb_accuracy_test11)
print("Accuracy Train: ", xgb_accuracy_train11)

print('LightGBM')
print("Accuracy Test: ", lgb_accuracy_test11)
print("Accuracy Train: ", lgb_accuracy_train11)
```

#### Análisis de modelos por cambios en hiperparámetros. Prueba 2 vs. Prueba 3
Por modelo

AdaBoost:
- Implementar learning_rate a 0.5: mejora el accuracy de test y train.

Gradient Boosting: 
- Learning_rate=0.05: empeora.
- Implementar Max depth a 2: se reduce el accuracy del test, asegurando evitar el sobreajuste.
- Implementar Subsample a 0.8: no se usa, empeora. 
- Implementar min_samples_leaf a != 1: no se usa, empeora.

XGBoost:
- Learning_rate=0.05: empeora.
- Implementar Max depth a 3: aumenta el accuracy test y se reduce el train, mejorando la métrica y reduciendo sobreajuste.
- Implementar Subsample a 0.8: se reduce mínimamente el accuracy test, no se usa. 
- Implementar colsample_bytree a 0.8: aumenta en 0.01 el accuracy test y el train. Mejora la métrica.
- Implementar Reg_lambda o reg_alpha a 0.1-0.8: no se usa, empeora. 

LightGBM:
- Implementar num_leaves a 20: mejora el accuracy test.
- Implementar Subsample a 0.8: no se usa, no mejora.
- Implementar Max depth: no se usa, no mejora.
- Implementar colsample_bytree a 0.7: mejora el accuracy test y reduce el train, mejorando la métrica y reduciendo sobreajuste.




#### Conclusión
Tras el ajuste de hiperparámetros, se observa una mejora notable en el rendimiento de los modelos basados en técnicas de boosting frente a los métodos tradicionales como el árbol de decisión y el Random Forest. En concreto, los modelos AdaBoost, Gradient Boosting, XGBoost y LightGBM alcanzan precisiones en el conjunto de test superiores al 90 %, superando ampliamente las obtenidas por los modelos iniciales.

Entre ellos, XGBoost destaca como el modelo con mayor precisión global (92,5 %), mostrando un excelente equilibrio entre sesgo y varianza, lo que lo convierte en la mejor opción si se busca maximizar el rendimiento predictivo. Por otro lado, AdaBoost presenta un comportamiento especialmente estable, con una diferencia mínima entre las métricas de entrenamiento y test (0,9 %), lo que refleja una gran capacidad de generalización y un bajo riesgo de sobreajuste. El modelo Gradient Boosting mantiene también un rendimiento muy competitivo y se consolida como una alternativa robusta y más fácilmente interpretable.

En conjunto, los resultados confirman que las técnicas de boosting, y en particular XGBoost, son las más adecuadas para este problema de clasificación binaria de diabetes, combinando alta precisión, estabilidad y adaptabilidad a las características del conjunto de datos empleado (668 registros, 6 variables predictoras y 1 objectivo).
